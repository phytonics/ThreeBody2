```python
from tris import ideal_pipeline
import matplotlib.pyplot as plt, cv2
plt.rcParams.update({'font.size': 22})
```

```python
def plot_and_export(kic):
    oc, _ = ideal_pipeline(f"../data/combined/kplr00{kic}.fits")
    oc["min_res"] = oc.residuals * 24 * 60
    oc.plot.scatter(
        x="time", y="min_res", figsize=(12, 8), #color="black",
        xlabel="Time / days", ylabel="O-C / min", legend=False,
        s=50
    )
    plt.savefig(f"{kic}.png", dpi=500)
    img = cv2.imread(f"{kic}.png")[400:3960, 150:5450, :]
    cv2.imwrite(f"{kic}.png", img)
    plt.imshow(img)
```

```python
plot_and_export(5384802)
```
